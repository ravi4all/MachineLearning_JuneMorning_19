```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10,100)
y = np.random.randint(20,2000,90)
```

```python
len(x),len(y)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y,s=60,marker='+')
```

```python
plt.plot(x,y,'ro')
```

```python
votes = [350,110,30,50]
parties = ['BJP','CONG','SP+BSP','Others']
```

```python
plt.pie(x=votes,labels=parties,startangle=90,counterclock=False,
       autopct='%1.1f%%',colors=['orange','yellow','red','gray'],explode=[0,0,0.3,0])
plt.legend(loc=0)
plt.title("2019 Election Results")
```

```python
from matplotlib import style
# style.use('ggplot')
style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y)
plt.savefig('plot_1.png')
```
